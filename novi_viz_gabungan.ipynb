```python
import pandas as pd
import matplotlib.pyplot as plt
from decimal import Decimal
import numpy as np

class CustomData:
    def __init__(self, url, columns):
        self.url = url
        self.columns = columns
        self.df = pd.read_csv(self.url, header=None)
        self.df = pd.DataFrame(self.df)
        self.process_data()

    def process_data(self):
        self.df[0] = self.df[0].apply(lambda x: x.split())
        self.df[0] = self.df[0].apply(lambda x: [Decimal(angka) for angka in x])
        self.df[0] = self.df[0].apply(self.list_to_string)
        self.df = self.df[0].str.split(' ', expand=True)
        self.df.columns = self.columns
        self.df = self.df.apply(lambda x: [Decimal(angka) for angka in x])

    def list_to_string(self, lst):
        return ' '.join(map(str, lst))

    def plot_data(self):
        plt.figure(figsize=(20, 5))
        for i, kolom in enumerate(self.columns):
            plt.subplot(1, len(self.columns), i+1)
            plt.plot(self.df['RHBO'], self.df[kolom], label=kolom)
            plt.xticks(np.arange(10))
            plt.xlabel('RHBO')
            plt.legend()
        plt.show()
```

```python
# Menggunakan class CustomData dengan dataset 'EOS_NSM_FSU2HZ2.dat' dan kolom 'RHBO', 'RHO', 'ED', 'BE'
dataset_url = 'https://raw.githubusercontent.com/rozi-bb/Dataset/main/EOS_NSM_FSU2HZ2.dat'
kolom_dataset = ['RHBO', 'RHO', 'ED', 'BE']
custom_data = CustomData(dataset_url, kolom_dataset)

# Menampilkan plot dari dataset yang sudah diproses
custom_data.plot_data()
```
